```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_csv('cars.csv')
```

```python
car
```

```python
sns.barplot(x="EngineSize",y="Horsepower",data=car)
plt.xticks(
    rotation=90, 
    horizontalalignment='right'
)
sns.set({'figure.figsize':(30,8)})
```

```python
sns.pointplot(x="Make",y="Horsepower",hue="Type",data=car)
plt.xticks(
    rotation=90, 
    horizontalalignment='right'
)
plt.gcf().set_size_inches(15, 8)
```

```python
sns.barplot(x="EngineSize",y="Horsepower",data=car)
plt.xticks(
    rotation=90, 
    horizontalalignment='right'
)
plt.gcf().set_size_inches(20, 9)
```
